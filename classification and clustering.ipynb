#### 3. Using Decision Tree

### Clustering

#### 1.Using K means

```python
from sklearn.cluster import KMeans
```

```python

clusterNum= 2
k_means=KMeans(init= 'k-means++',n_clusters=clusterNum,n_init=12) #here k-means++ means its a method to choose initial cluster centeroids
# n_init means how many times we need to move the centeroids
k_means.fit(X)
labels = k_means.labels_
labels # model predicted
```

```python
k_means.cluster_centers_ # its a parameter to find coordinates of each centroids according to the features
# here we have four features and 3 k values so it has 3 centroids
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg= silhouette_score(X,labels)
silhouette_avg
```

```python
# suppose we use multiple values
```

```python
Nc=range(1,20)
kmeans=[KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))] # list comprehension
score
```

```python
plt.plot(Nc,score)
plt.xlabel('Number of clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
clusterNum= 3
k_means=KMeans(init= 'k-means++',n_clusters=clusterNum,n_init=12) 
k_means.fit(X)
labels = k_means.labels_
labels
k_means.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg= silhouette_score(X,labels)
silhouette_avg
```

```python
#here silhoutte value is o.5 which is close to 1, so it considered as a good clustering solution.
```

#### 2. Using Hierarcial clustering 

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
#plt.grid(True)
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_ag=ag.fit_predict(X)
y_ag
```

```python
# the above output has 3 clusters.
# so now we will plot this by using scatter plot 
```

```python
plt.figure(figsize=(12,7))
plt.scatter(X[y_ag == 0, 0], X[y_ag == 0, 1], s = 100, c = 'red', label = 'setosa')
plt.scatter(X[y_ag == 1, 0], X[y_ag == 1, 1], s = 100, c = 'blue', label = 'virginca')
plt.scatter(X[y_ag == 2, 0], X[y_ag == 2, 1], s = 100, c = 'green', label = 'versicolor')

plt.legend(fontsize=16)
plt.grid(True)

plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
[53]
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
