### Data Preprocessing

```python
dummies=list(iris.columns[iris.dtypes=='object'])
dummies
```

```python

import numpy as np
df_one_hot_encoded = pd.get_dummies(data = iris, columns = dummies)

print(df_one_hot_encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

cols_to_encode = dummies
new_encoded_cols_names = []

for col in cols_to_encode:
    new_encoded_cols_names += [f"is_{category}" for category in iris[col].unique().tolist()]

print(new_encoded_cols_names)


one_hot_encoder = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
encoded_cols = one_hot_encoder.fit_transform(iris[cols_to_encode])


df_encoded = pd.DataFrame(encoded_cols, columns = new_encoded_cols_names)
df_one_hot_encoded = iris.join(df_encoded)

print(df_one_hot_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in iris:
    iris[i] = label_encoder.fit_transform(iris[i])
```

### Modeling

```python
from sklearn.datasets import load_iris
%matplotlib inline
import matplotlib.pyplot as plt
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.data[0]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target, test_size=0.3)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(iris.data)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
